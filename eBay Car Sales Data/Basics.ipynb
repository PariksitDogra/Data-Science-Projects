```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'ab_test',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'num_photos', 'postal_code',
       'last_seen']
```

I replaced the camelcase names of the columns with snakecase to better align it to our use in Python. Furthermore, I renamed some columns to better represent the columns values.

```python
autos.head()
```

```python
autos.describe(include='all')
autos = autos.drop(["num_photos", "seller", "offer_type"], axis=1)
```

As can be seen, price and odometer can be converted to be purely numeric.

```python
autos["price"] = autos["price"].str.replace("$", "").str.replace(",","")
autos["price"] = autos["price"].astype(int)
```

```python
autos["odometer"] = autos["odometer"].str.replace("km","").str.replace(",","")

autos.rename({"odometer":"odometer_km"}, axis=1, inplace=True)
```

```python
autos["odometer_km"] = autos["odometer_km"].astype(int)
autos["odometer_km"].unique()
```

```python
brand_counts = autos['brand'].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > 0.05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_price = brand_only["price"].mean()
    
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

Of the top 5 brands, there is a distinct price gap:

Audi, BMW and Mercedes Benz are more expensive
Ford and Opel are less expensive
Volkswagen is in between - this may explain its popularity, it may be a 'best of 'both worlds' option.

```python
brand_mean_mileage = {}
for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_mileage = brand_only['odometer_km'].mean()
    
    brand_mean_mileage[brand] = int(mean_mileage)
brand_mean_mileage
```

```python
bmp = pd.Series(brand_mean_prices)
bmm = pd.Series(brand_mean_mileage)

brand_info = pd.DataFrame(bmp, columns=['mean_price'])
brand_info
```
